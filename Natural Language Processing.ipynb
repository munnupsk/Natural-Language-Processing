# Stemming

```python
  # stemming is a process of reducing infected or derived words 
  # to thier word stem,base or root form
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
#stopwords is used to remove the kind of words like the,and,or,they....etc
```

```python
stopwords.words('english')
```

```python
#PorterStemmer is used for stemming purpose
```

```python
stemmer=PorterStemmer()
```

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=" ".join(words)
```

# Lemmatization

```python
# for example if we lemmatize word "integence" then we get intelligent
#In the stemming it was not the case
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
from nltk.corpus import stopwords
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=" ".join(words)
```

## Cleaning the text

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
wordnet=WordNetLemmatizer()
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
corpus=[]
```

```python
for i in range(len(sentences)):
    review=re.sub('[^a-zA-Z]',' ',sentences[i])# this particular code while remove all ,.! etc and replces them with space
    review=review.lower()
    review=review.split()
    review=[wordnet.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Creating a bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
cv=CountVectorizer()
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(X)
```
